```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

```

```python

x = np.linspace(0, 1, 100)

P = np.linspace(0.5, 1, 6)

# Y = p * np.log (1 + m) − (1 − p) log (1 − m)
for p in P:
  plt.plot(x, p * np.log(1 + x) + (1 - p) * np.log(1 - x), label=f'p = {p}')

# plt.legend(
#   loc='upper center',
#   bbox_to_anchor=(0.5, 1.15),
#   ncol=3,
# )

# add legend to left of plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# axis labels
plt.xlabel('Bet Fraction of original value')
plt.ylabel('Delta in Expected Utility (log scale')


# set y scale to log
plt.yscale('log')

plt.grid()
plt.xticks(np.arange(0, 1.1, 0.1))
# plt.yticks(np.arange(10**(-4), 1, 10**(-1)))
plt.title("Expected Utility of Betting on a Coin Toss")
plt.show()



```
